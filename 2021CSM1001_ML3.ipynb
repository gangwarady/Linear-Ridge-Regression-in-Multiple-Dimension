```python
def predridgereg(X, weights):
  y = numpy.matmul(X,weights)

  return  y
```

```python
def ridgereg(X, y, weights, learning_rate, lam):
  y_pred = predridgereg(X, weights)

  y=y.reshape((len(y),1))
  
  cost = (numpy.sum((y_pred-y)**2))/2 + (lam*(numpy.sum((weights)**2)))/2
 
  count=0
  weights = weights - learning_rate*( numpy.matmul(X.T,(y_pred - y)) + lam*(weights))
  #print(weights.shape)
  y_pred = predridgereg(X, weights)
  new_cost =  (numpy.sum((y_pred-y)**2))/2 + (lam*(numpy.sum((weights)**2)))/2
  iterations = 100
 
  while(new_cost<cost):
    count = count+1
    cost = new_cost
    
    weights = weights - learning_rate*( numpy.matmul(X.T,(y_pred - y)) + lam*(weights))
    y_pred = predridgereg(X, weights)
    new_cost =  (numpy.sum((y_pred-y)**2))/2 + (lam*(numpy.sum((weights)**2)))/2
  
  #print(cost)
  #print(new_cost)
  #print(count)
  return cost,weights,y_pred
    
```

**Observations :**


*   Both ridgereg() and preridgereg() function have been implemented above.
*   preridgereg() function simply returns y prediction values by performing matrix multiplication of input X with the weights w.


*   ridgereg() functions performs gradient descent where it keeps searching for efficient w and its uses regularization for avoiding overfitting. It return minimum cost, efficient/best wieghts and y predictions wrt to the efficient/best weights.


* Learning rate value used is = 0.01










```python
def k_fold(data, k, lam):
  fold_size = int(len(data)/k)
  valida = data
  Ew = []
  w = numpy.random.randn(7,1)
  weigh = []
  for i in range(k):
    train = data
    vai = valida.sample(n=fold_size, random_state= None)
    valida = valida.drop(vai.index)
    train = train.drop(vai.index)
    #print(train)
    X = []
    y_train= train['charges']
    y_train = numpy.array(y_train)
    for ind in train.index:
      row=[]
      f = True
      while(f==True):
        row.append(1)
        row.append(train['age'][ind])
        row.append(train['sex'][ind])
        row.append(train['bmi'][ind])
        row.append(train['children'][ind])
        row.append(train['smoker'][ind])
        row.append(train['region'][ind])
        f=False
      X.append(row)
    X = numpy.array(X)
    #print(X)
   
    #print(w.shape)
    learning_rate=0.01
    c,wei,y_return=ridgereg(X, y_train, w, learning_rate, lam = lam)
    weigh.append(wei)
    Ew.append(c)
   
  return Ew, weigh


```



*   The above function is the k-fold function where it divides the training validation data into 10 parts and always sends the different 9 training parts to the ridgereg() function.
*   It creates the Matrix Form for Multivariate Regression for our training data and then calls the ridgereg() function using it.


*   It finally returns the weights and cost/loss value for each of the folds.






```python
lamda_t = [*range(-10, 10, 1) ]

fig, axs = plt.subplots(2,10, figsize=(20, 10))
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()
count=0
train_erms = []
eff_loss = 100
m_l = 100
wstar = []
for lval in lamda_t:
  loss, wstar = k_fold(train_val, 10, lval)
  #print(loss)
  loss = numpy.array(loss)
  k = numpy.min(loss)
  errw = numpy.sqrt((2*k)/963)
  train_erms.append(errw)
  if(errw<eff_loss):
    
    ind = numpy.argmin(loss)
    eff_loss=errw 
    m_l=lval
    eff_w = wstar[ind]
  erms = []
  for l in loss:
    erms.append(numpy.sqrt((2*l)/train_val.shape[0]))
  
  k_values = [*range(1, 11, 1)]
  axs[count].plot(k_values, erms)
  axs[count].set_title("lamda = %i " %lval)
  count = count + 1
```

**Observations:**


*   The above subplots plot different root mean square error for each fold and for each lambda value.
*   For each fold of particular lamda value we get different root mean square erros.


*   The value of k in k-fold is 10.
*   The subplots have been plotted for different lambda values ranging from -10 to 9.





```python

print("Lamda Value = " ,m_l)
print(eff_w)
plt.plot(lamda_t,train_erms, c = 'blue')
plt.legend(['ERMS'])
```


**Observations:**
*  In order to know which lambda should be chosen, I have taken the minimum root mean square error for each lambda and plotted the above graph.
*   The above graph shows different root mean square errors for different values of lambda. 


*   To find the lambda I found out the minimum root mean square error(Erms) out of all the root mean square errors for each fold of each lambda value. So the lambda vaule which got is -10.
*   The above graph which shows the minimum Erms for different values of lambda also shows that we get minimum Erms at lambda value -10.





```python
train_error = []
test_error = []
y_train_val = predridgereg(train_val, eff_w)
y_test = predridgereg(test, eff_w)
y_train_true = numpy.array(train_val['charges'])
y_train_true= y_train_true.reshape((len(y_train_true),1))
y_test_true = numpy.array(test['charges'])
y_test_true= y_test_true.reshape((len(y_test_true),1))


for lval in lamda_t:
  loss_train_val = (numpy.sum((y_train_val-y_train_true)**2))/2 + (lval*(numpy.sum((eff_w)**2)))/2
  train_error.append(numpy.sqrt((2*loss_train_val)/1070))
  
  loss_test = (numpy.sum((y_test-y_test_true)**2))/2 + (lval*(numpy.sum((eff_w)**2)))/2
  test_error.append(numpy.sqrt((2*loss_test)/268))

#diff_y = (y_train_val - y_test)
#print(diff_y)
#variance = numpy.var(diff_y)
#plt.plot(lamda_t, variance, c="green")
train_error = numpy.array(train_error)
test_error = numpy.array(test_error)
plt.plot(lamda_t,train_error, c="blue")
plt.plot(lamda_t,test_error, c="red")
plt.legend(['Training error', 'Test error'])
test_error__mean = numpy.mean(test_error)
bias_square = (train_error-test_error__mean)**2 
variance = abs(train_error - test_error)
```

**Observations:**


*   The above graph shows Erms(root mean square error) value for training and testing data for different value of lambda.
*   We can see that the testing error is always greater than the training error.


*  The training and testing error increase as we increase the lambda value.
*  The testing error increases more significantly than the training error when we increase the value of lambda. This leads to increase in variance as we increase lambda.
* At lamda = -10 the training error and testing error are close to each other. So the bias is high at lesser value of lambda and bias decreases as we increase the value of lambda.
